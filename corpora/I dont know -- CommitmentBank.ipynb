```python
import re
import pandas as pd
```

```python
def _extract(df, conds):    
    cond = []
    for k, v in conds.items():
        cond.append(df[k] == v)

    cond = list(map(all, zip(*cond)))
    out = df[cond]
    return out
```

```python
df = pd.read_csv("CommitmentBank-items.csv")
```

```python
conds = {"Verb": "know", "Embedding": "negation", 
         "MatTense": "present", "MatSubjLemma": "I"}

sub = _extract(df, conds)
```
